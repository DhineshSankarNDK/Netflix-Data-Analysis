```python
# =====================================================
# NETFLIX DATA ANALYSIS PROJECT
# =====================================================

# -------------------------------
# IMPORT LIBRARIES
# -------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# VISUALIZATION SETTINGS
# -------------------------------

sns.set_theme(style="whitegrid")

# -------------------------------
# LOAD DATASET
# -------------------------------

df = pd.read_csv("../data/netflix_titles.csv")

# -------------------------------
# BASIC DATA EXPLORATION
# -------------------------------

print("First 5 Rows")
print(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nDataset Information:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nStatistical Summary:")
print(df.describe())

# =====================================================
# DATA CLEANING
# =====================================================

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill missing values
df['country'] = df['country'].fillna('Unknown')
df['director'] = df['director'].fillna('Not Available')
df['cast'] = df['cast'].fillna('Not Available')
df['rating'] = df['rating'].fillna('Unknown')

# Convert date column
df['date_added'] = pd.to_datetime(
    df['date_added'],
    errors='coerce'
)

# Create year_added column
df['year_added'] = df['date_added'].dt.year

print("\nData Cleaning Completed")

# =====================================================
# MOVIES VS TV SHOWS
# =====================================================

type_counts = df['type'].value_counts()

plt.figure(figsize=(8, 8))

plt.pie(
    type_counts.values,
    labels=type_counts.index,
    autopct='%1.1f%%',
    wedgeprops={'width': 0.4},
    startangle=90
)

plt.title(
    "Netflix Content Distribution",
    fontsize=18,
    fontweight='bold'
)

plt.tight_layout()

# Save image
plt.savefig("../images/content_distribution.png")

plt.show()

# =====================================================
# TOP 10 GENRES
# =====================================================

top_genres = (
    df['listed_in']
    .value_counts()
    .head(10)
)

plt.figure(figsize=(14, 7))

ax = sns.barplot(
    x=top_genres.values,
    y=top_genres.index
)

# Add labels
for i, v in enumerate(top_genres.values):
    ax.text(
        v + 10,
        i,
        str(v),
        va='center',
        fontsize=11,
        fontweight='bold'
    )

plt.title(
    "Top 10 Genres on Netflix",
    fontsize=20,
    fontweight='bold'
)

plt.xlabel("Number of Titles", fontsize=13)
plt.ylabel("Genres", fontsize=13)

plt.tight_layout()

plt.savefig("../images/top_genres.png")

plt.show()

# =====================================================
# RELEASE YEAR TREND
# =====================================================

release_year = (
    df['release_year']
    .value_counts()
    .sort_index()
)

plt.figure(figsize=(16, 7))

plt.plot(
    release_year.index,
    release_year.values,
    marker='o',
    linewidth=3
)

# Highlight peak year
peak_year = release_year.idxmax()
peak_value = release_year.max()

plt.scatter(
    peak_year,
    peak_value,
    s=200
)

plt.text(
    peak_year,
    peak_value + 20,
    f'Peak: {peak_year}',
    ha='center',
    fontsize=12,
    fontweight='bold'
)

plt.title(
    "Netflix Content Release Trend",
    fontsize=20,
    fontweight='bold'
)

plt.xlabel("Release Year", fontsize=13)
plt.ylabel("Number of Titles", fontsize=13)

plt.grid(True)

plt.tight_layout()

plt.savefig("../images/release_trend.png")

plt.show()

# =====================================================
# HEATMAP
# =====================================================

heatmap_data = pd.crosstab(
    df['year_added'],
    df['type']
)

plt.figure(figsize=(12, 8))

sns.heatmap(
    heatmap_data,
    annot=True,
    fmt='d',
    cmap='YlGnBu'
)

plt.title(
    "Content Added by Year and Type",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel("Content Type")
plt.ylabel("Year Added")

plt.tight_layout()

plt.savefig("../images/heatmap.png")

plt.show()

# =====================================================
# RATINGS DISTRIBUTION
# =====================================================

plt.figure(figsize=(14, 7))

rating_order = (
    df['rating']
    .value_counts()
    .index
)

ax = sns.countplot(
    y='rating',
    data=df,
    order=rating_order
)

# Add labels
for p in ax.patches:
    ax.annotate(
        str(int(p.get_width())),
        (
            p.get_width(),
            p.get_y() + p.get_height()/2
        ),
        ha='left',
        va='center',
        fontsize=10
    )

plt.title(
    "Netflix Ratings Distribution",
    fontsize=20,
    fontweight='bold'
)

plt.xlabel("Count")
plt.ylabel("Rating")

plt.tight_layout()

plt.savefig("../images/ratings_distribution.png")

plt.show()

# =====================================================
# TOP COUNTRIES
# =====================================================

top_countries = (
    df['country']
    .value_counts()
    .head(10)
)

plt.figure(figsize=(15, 7))

ax = sns.barplot(
    x=top_countries.values,
    y=top_countries.index
)

for i, v in enumerate(top_countries.values):
    ax.text(
        v + 5,
        i,
        str(v),
        va='center',
        fontsize=11,
        fontweight='bold'
    )

plt.title(
    "Top Countries Producing Netflix Content",
    fontsize=20,
    fontweight='bold'
)

plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.tight_layout()

plt.savefig("../images/top_countries.png")

plt.show()

# =====================================================
# HISTOGRAM
# =====================================================

plt.figure(figsize=(15, 6))

sns.histplot(
    df['release_year'],
    bins=30,
    kde=True
)

plt.title(
    "Distribution of Netflix Release Years",
    fontsize=20,
    fontweight='bold'
)

plt.xlabel("Release Year")
plt.ylabel("Frequency")

plt.tight_layout()

plt.savefig("../images/release_histogram.png")

plt.show()

# =====================================================
# BOXPLOT
# =====================================================

plt.figure(figsize=(10, 5))

sns.boxplot(
    x=df['release_year']
)

plt.title(
    "Boxplot of Release Years",
    fontsize=18,
    fontweight='bold'
)

plt.tight_layout()

plt.savefig("../images/release_boxplot.png")

plt.show()

# =====================================================
# FINAL INSIGHTS
# =====================================================

print("\n====================================")
print("NETFLIX DATA ANALYSIS - INSIGHTS")
print("====================================")

print("\n1. Netflix has significantly more Movies than TV Shows.")

print("\n2. Drama and International genres dominate Netflix.")

print("\n3. Content production increased rapidly after 2015.")

print("\n4. TV-MA is one of the most common content ratings.")

print("\n5. The United States contributes the highest amount of content.")

print("\n6. Netflix experienced peak growth in recent years.")

print("\n7. Heatmap analysis shows strong growth in both Movies and TV Shows.")

print("\nProject Completed Successfully.")
```
